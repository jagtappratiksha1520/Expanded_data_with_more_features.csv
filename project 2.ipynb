# distribution of ethnic groups

```python
import matplotlib.pyplot as plt

groupA = data.loc[data['EthnicGroup'] == "group A"].count()["EthnicGroup"]
groupB = data.loc[data['EthnicGroup'] == "group B"].count()["EthnicGroup"]
groupC = data.loc[data['EthnicGroup'] == "group C"].count()["EthnicGroup"]
groupD = data.loc[data['EthnicGroup'] == "group D"].count()["EthnicGroup"]
groupE = data.loc[data['EthnicGroup'] == "group E"].count()["EthnicGroup"]

l = ["groupA", "groupB", "groupC", "groupD", "groupE"]
mlist = [groupA, groupB, groupC, groupD, groupE]
plt.pie(mlist, labels=l,autopct ="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
